```python
def calculate_balls(over):
    try:
        if isinstance(over, str):
            # Split the string into two parts
            over_parts = over.split('.')
            first_part = int(over_parts[0])
            second_part = int(over_parts[1]) if len(over_parts) > 1 else 0
        else:
            # If it's not a string, assume it's already a valid integer
            first_part = int(over)
            second_part = ( over -int(over))*10
        
        total_balls = (first_part * 6) + second_part
    except:
        print(over,type(over))
    return int(total_balls)

# Apply the function to the 'over' column and create a new 'balls' column
df['balls'] = df['overs'].apply(calculate_balls)
```
